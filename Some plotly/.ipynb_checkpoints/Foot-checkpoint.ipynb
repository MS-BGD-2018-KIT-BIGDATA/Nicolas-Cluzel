```python
def getSoupFromURL(url):
    # Gets the soup from the input url:
    
    res = requests.get(url)
    if res.status_code == 200:
        return BeautifulSoup(res.text, 'html.parser')
```

```python
def getGoals(player_surname, player_name):
    # Gets the total goals of a specific player:
        
    name = [] 
    seasons = []
    years = []
    clubs = []
    championship_games = []
    championship_goals = []   
    european_games = []
    european_goals = []
    international_games = []
    international_goals = []
    goals_list = []
    features = ['seasons', 'clubs', 
                'championship_games', 'championship_goals',
                'european_games', 'european_goals',
                'international_games', 'international_goals'
                ]
    synthesis = pd.DataFrame()
    url = "http://www.sofoot.com/" + player_surname + "-" + player_name + ".html"
    soup = getSoupFromURL(url)
    
    #    try:
    #gotdata = dlist[1]
    #except IndexError:
    #gotdata = 'null'
    
    if soup:
        for i in range (100):
            try:
                row = soup.find_all("tr", class_="carriere-ligne fst-line")[i]
                seasons.append(
                        row.find_all("td", class_ = "carriere-ligne-item carriere-saison")[0].text.strip())
                years.append(
                        row.find_all("td", class_ = "carriere-ligne-item carriere-saison")[0].text.strip().split('/')[0])
                clubs.append(
                        row.find_all("td", class_ = "carriere-ligne-item carriere-club")[0].text.strip())
                championship_games.append(
                        row.find_all("td", class_ = "carriere-ligne-item carriere-champ-matchs")[0].text.strip())
                championship_goals.append(
                        row.find_all("td", class_ = "carriere-ligne-item carriere-champ-buts")[0].text.strip())
                european_games.append(
                        row.find_all("td", class_ = "carriere-ligne-item carriere-cont-matchs")[0].text.strip())
                european_goals.append(
                        row.find_all("td", class_ = "carriere-ligne-item carriere-cont-buts")[0].text.strip())
                international_games.append(
                        row.find_all("td", class_ = "carriere-ligne-item carriere-int-matchs")[0].text.strip())
                international_goals.append(
                        row.find_all("td", class_ = "carriere-ligne-item carriere-int-buts")[0].text.strip())
            except IndexError:
                row = 'null'
        
    for j in range(len(seasons)):
        name.append(player_surname.capitalize() + " " + player_name.capitalize())
    
    synthesis['Player'] = name
    synthesis['Season'] = seasons
    synthesis['Club'] = clubs
    synthesis['Championship_games'] = championship_games
    synthesis['Championship_goals'] = championship_goals
    synthesis['European_games'] = european_games
    synthesis['European_goals'] = european_goals
    synthesis['International_games'] = international_games
    synthesis['International_goals'] = international_goals
    synthesis['Year'] = years
    synthesis[['Championship_games', 
               'Championship_goals', 
               'European_games',
               'European_goals',
               'International_games',
               'International_goals', 'Year']] = synthesis[['Championship_games', 
                                      'Championship_goals', 
                                      'European_games', 
                                                    'European_goals', 
                                                    'International_games', 
                                                    'International_goals', 'Year']].apply(pd.to_numeric, errors='raise')
    synthesis['Total_club_goals'] = synthesis['European_goals'] + synthesis['Championship_goals']
    
    #if synthesis['Championship_games'].any() != 0:
    synthesis['Championship_ratio'] = synthesis['Championship_goals'] / synthesis['Championship_games']
    #elif synthesis['Championship_games'].any() == 0:
    #    synthesis['Championship_ratio'] =0
    synthesis['European_ratio'] = synthesis['European_goals'] / synthesis['European_games']
    #if synthesis['International_games'].any() != 0:
    #if synthesis['International_games'].notna():
    #if pd.notnull(synthesis['International_games']):
    #synthesis['International_ratio'] = 0
    synthesis['International_ratio'] = synthesis['International_goals'] / synthesis['International_games']
    #synthesis['Test'] = 1
    #synthesis['Test'] = synthesis['International_games'].loc[synthesis['International_games'] >= 11]
    #synthesis['Test'] = synthesis['International_games'] >= 11
    #synthesis['International_ratio'].apply(lambda x: 0 if(np.all(pd.isnull(x[1]))) else x, axis = 1)
    #elif synthesis['International_games'].isnull():
    #if pd.isnull(synthesis['International_games']):
    #    synthesis['International_ratio'] = 0
    synthesis = synthesis.fillna(0)                                                                       
        
    synthesis.set_index('Player', inplace=True)
        
    return synthesis
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_ebola.csv')
df.head()

cases = []
colors = ['rgb(239,243,255)','rgb(189,215,231)','rgb(107,174,214)','rgb(33,113,181)']
months = {6:'June',7:'July',8:'Aug',9:'Sept'}

for i in range(6,10)[::-1]:
    cases.append(go.Scattergeo(
            lon = df[ df['Month'] == i ]['Lon'], #-(max(range(6,10))-i),
            lat = df[ df['Month'] == i ]['Lat'],
            text = df[ df['Month'] == i ]['Value'],
            name = months[i],
            marker = dict(
                size = df[ df['Month'] == i ]['Value']/50,
                color = colors[i-6],
                line = dict(width = 0)
            )
        )
    )

cases[0]['text'] = df[ df['Month'] == 9 ]['Value'].map('{:.0f}'.format).astype(str)+' '+\
    df[ df['Month'] == 9 ]['Country']
cases[0]['mode'] = 'markers+text'
cases[0]['textposition'] = 'bottom center'

inset = [
    go.Choropleth(
        locationmode = 'country names',
        locations = df[ df['Month'] == 9 ]['Country'],
        z = df[ df['Month'] == 9 ]['Value'],
        text = df[ df['Month'] == 9 ]['Country'],
        colorscale = [[0,'rgb(0, 0, 0)'],[1,'rgb(0, 0, 0)']],
        autocolorscale = False,
        showscale = False,
        geo = 'geo2'
    ),
    go.Scattergeo(
        lon = [21.0936],
        lat = [7.1881],
        text = ['Africa'],
        mode = 'text',
        showlegend = False,
        geo = 'geo2'
    )
]

layout = go.Layout(
    title = 'Ebola cases reported by month in West Africa 2014<br> \
Source: <a href="https://data.hdx.rwlabs.org/dataset/rowca-ebola-cases">\
HDX</a>',
    geo = dict(
        resolution = 50,
        scope = 'africa',
        showframe = False,
        showcoastlines = True,
        showland = True,
        landcolor = "rgb(229, 229, 229)",
        countrycolor = "rgb(255, 255, 255)" ,
        coastlinecolor = "rgb(255, 255, 255)",
        projection = dict(
            type = 'Mercator'
        ),
        lonaxis = dict( range= [ -15.0, -5.0 ] ),
        lataxis = dict( range= [ 0.0, 12.0 ] ),
        domain = dict(
            x = [ 0, 1 ],
            y = [ 0, 1 ]
        )
    ),
    geo2 = dict(
        scope = 'africa',
        showframe = False,
        showland = True,
        landcolor = "rgb(229, 229, 229)",
        showcountries = False,
        domain = dict(
            x = [ 0, 0.6 ],
            y = [ 0, 0.6 ]
        ),
        bgcolor = 'rgba(255, 255, 255, 0.0)',
    ),
    legend = dict(
           traceorder = 'reversed'
    )
)

fig = go.Figure(layout=layout, data=cases+inset)
iplot( fig, validate=False, filename='West Africa Ebola cases 2014' )
```

```python
import plotly.plotly as py
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_us_cities.csv')
df.head()

df['text'] = df['name'] + '<br>Population ' + (df['pop']/1e6).astype(str)+' million'
limits = [(0,2),(3,10),(11,20),(21,50),(50,3000)]
colors = ["rgb(0,116,217)","rgb(255,65,54)","rgb(133,20,75)","rgb(255,133,27)","lightgrey"]
cities = []
scale = 5000

for i in range(len(limits)):
    lim = limits[i]
    df_sub = df[lim[0]:lim[1]]
    city = dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = df_sub['lon'],
        lat = df_sub['lat'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['pop']/scale,
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1]) )
    cities.append(city)

layout = dict(
        title = '2014 US city populations<br>(Click legend to toggle traces)',
        showlegend = True,
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)",
            countrycolor="rgb(255, 255, 255)"
        ),
    )

fig = dict( data=cities, layout=layout )
iplot( fig, validate=False, filename='d3-bubble-map-populations' )
```
